```python
nominal_category=['gender','Partner','Dependents','PaymentMethod']
ordinal_category=['PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup',
                 'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','Contract','PaperlessBilling']
```

```python
Categorical_data=nominal_category+ordinal_category
```

```python
Numerical_data = [(i) for i in X.columns if i not in Categorical_data]
```

```python
Numerical_data
```

```python
X.head()
```

```python
for i in nominal_category:
    
    X = pd.concat([X, pd.get_dummies(X[i],drop_first=True,prefix=i)], axis=1)
    
  
    
```

```python
X=X.drop(['gender','Partner','Dependents','PaymentMethod'], axis=1)
```

```python
X.head()
```

```python
import category_encoders as ce
import pandas as pd

# create object of Ordinalencoding
encoder= ce.OrdinalEncoder(cols=['PhoneService'],return_df=True,
                           mapping=[{'col':'PhoneService',
'mapping':{'No':0,'Yes':1}}])

X['PhoneService'] = encoder.fit_transform(X['PhoneService'])
#Original data
```

```python
e1= ce.OrdinalEncoder(cols=['MultipleLines'],return_df=True,
                           mapping=[{'col':'MultipleLines',
'mapping':{'No phone service':0,'No':1,'Yes':2}}])

X['MultipleLines'] = e1.fit_transform(X['MultipleLines'])
```

```python
e2= ce.OrdinalEncoder(cols=['InternetService'],return_df=True,
                           mapping=[{'col':'InternetService',
'mapping':{'No':0,'Fiber optic':1,'DSL':2}}])

X['InternetService'] = e2.fit_transform(X['InternetService'])
```

```python
e3= ce.OrdinalEncoder(cols=['OnlineSecurity'],return_df=True,
                           mapping=[{'col':'OnlineSecurity',
'mapping':{'No internet service':0,'No':1,'Yes':2}}])

X['OnlineSecurity'] = e3.fit_transform(X['OnlineSecurity'])
```

```python
e4= ce.OrdinalEncoder(cols=['OnlineBackup'],return_df=True,
                           mapping=[{'col':'OnlineBackup',
'mapping':{'No internet service':0,'No':1,'Yes':2}}])

X['OnlineBackup'] = e4.fit_transform(X['OnlineBackup'])
```

```python
e6= ce.OrdinalEncoder(cols=['DeviceProtection'],return_df=True,
                           mapping=[{'col':'DeviceProtection',
'mapping':{'No internet service':0,'No':1,'Yes':2}}])

X['DeviceProtection'] = e6.fit_transform(X['DeviceProtection'])
```

```python
e7= ce.OrdinalEncoder(cols=['TechSupport'],return_df=True,
                           mapping=[{'col':'TechSupport',
'mapping':{'No internet service':0,'No':1,'Yes':2}}])

X['TechSupport'] = e7.fit_transform(X['TechSupport'])
```

```python
e8= ce.OrdinalEncoder(cols=['StreamingTV'],return_df=True,
                           mapping=[{'col':'StreamingTV',
'mapping':{'No internet service':0,'No':1,'Yes':2}}])

X['StreamingTV'] = e8.fit_transform(X['StreamingTV'])
```

```python
e9= ce.OrdinalEncoder(cols=['StreamingMovies'],return_df=True,
                           mapping=[{'col':'StreamingMovies',
'mapping':{'No internet service':0,'No':1,'Yes':2}}])

X['StreamingMovies'] = e9.fit_transform(X['StreamingMovies'])
```

```python
e10= ce.OrdinalEncoder(cols=['Contract'],return_df=True,
                           mapping=[{'col':'Contract',
'mapping':{'Month-to-month':0,'One year':1,'Two year':2}}])

X['Contract'] = e10.fit_transform(X['Contract'])
```

```python
e12= ce.OrdinalEncoder(cols=['PaperlessBilling'],return_df=True,
                           mapping=[{'col':'PaperlessBilling',
'mapping':{'No':0,'Yes':1}}])

X['PaperlessBilling'] = e12.fit_transform(X['PaperlessBilling'])
```

```python
e13= ce.OrdinalEncoder(cols=['Churn'],return_df=True,
                           mapping=[{'col':'Churn',
'mapping':{'No':0,'Yes':1}}])

y['Churn'] = e13.fit_transform(y['Churn'])
```
